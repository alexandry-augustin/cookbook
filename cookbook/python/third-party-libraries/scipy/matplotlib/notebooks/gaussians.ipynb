```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy.stats
import numpy as np
```

```python
x_min = 0.0
x_max = 30.0
```

```python
# Population A
mean = 9.0 
std = 2.0

x1 = np.linspace(x_min, x_max, 100)
y1 = scipy.stats.norm.pdf(x1,mean,std)
```

```python
# Population B
mean = 15.0 
std = 4.0

x2 = np.linspace(x_min, x_max, 100)
y2 = scipy.stats.norm.pdf(x2, mean, std)
```

# Plot

```python
fig, ax = plt.subplots(1, 1)

ax.grid()

ax.plot(x1, y1, color='black')
ax.fill_between(x1, y1, color='#0b559f', alpha='1.0')

ax.plot(x2, y2, color='black')
ax.fill_between(x2, y2, color='#89bedc', alpha='1.0')

pop_a = mpatches.Patch(color='#0b559f', label='Population A')
pop_b = mpatches.Patch(color='#89bedc', label='Population B')

ax.legend(handles=[pop_a,pop_b])

plt.xlim(x_min,x_max)
plt.ylim(0,0.25)

plt.xlabel('x')
plt.ylabel('Probability Density Function')
```
